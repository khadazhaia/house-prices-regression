# Cleaning raw data to get ready for modeling

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv("/Users/sa17/Desktop/house-prices-regression/data/raw/train.csv")

df
```

```python
# Create a new dataframe with only the columns I want to predict SalePrice with. These columns are pulled from my previous knowdledge/ research on Real Estate and Multivariate Analysis

columns = ["SalePrice", "MSSubClass", "LotArea", "OverallQual", "GrLivArea", "GarageCars", "1stFlrSF", "FullBath", "TotRmsAbvGrd", "YearBuilt", "YearRemodAdd", "Foundation", "Fireplaces", "ExterQual", "Neighborhood"]

new_df = df[columns]

new_df
```

```python
# Checking for null values 
null_values = new_df.isna().sum()

null_values
```

```python
# Remove outliers in SalePrice. Gain insights from Sale Price Distrubution graph 
Q1 = new_df["SalePrice"].quantile(0.25)
Q3 = new_df["SalePrice"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

clean_df = new_df[(new_df["SalePrice"] >= lower_bound) & (new_df["SalePrice"] <= upper_bound)]

clean_df
```

```python
# One-hot encode all categorical columns 
cat_columns = clean_df.select_dtypes(include=["object"]).columns

encoder = OneHotEncoder(sparse_output=False, handle_unknown="ignore") 

# Fit, transform and create a new dataframe 
onehot = encoder.fit_transform(clean_df[cat_columns])

onehot_df = pd.DataFrame(onehot, columns=encoder.get_feature_names_out(cat_columns), index=clean_df.index)

# Add the one-hot encoded columns to the original DataFrame and drop original categorical columns
encoded_df = pd.concat([clean_df.drop(columns=cat_columns), onehot_df], axis=1)

encoded_df
```
